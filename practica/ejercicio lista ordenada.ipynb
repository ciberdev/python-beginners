```python
def ordenar(nums):
    # 1. por cada item en la lista (con indice i ascendente):
    # 2. comparar item[i] con item[i+1]
    # 3. si item[1] > item[i+1] entonces intercambiar
    # 4. sino, seguir de largo.
    
    # Mi variable de control de intercambios
    huboIntercambio = True
    #....
    while huboIntercambio:
        huboIntercambio = False
        # ...
        for i in range(len(nums)-1):
            if nums[i] < nums[i + 1]:
                nums[i], nums[i+1] = nums[i+1], nums[i]
                huboIntercambio = True

#Vamos a ordenar esta lista.
lista = [2, -42, 20, 30, 1, 7, 25, 11, 8, 16]
ordenar(lista)
print(lista)

                
```

```python
with open("test.csv", "r") as data:
    nombres = []
    for row in data:
        campos = row.split(",")
        # Agregamos el valor del campo de nombre a un nuevo arreglo (nombres)
        nombres.append(campos[1])        
    
    for nombre in nombres:
        # utilizamos la funcion count sobre el arreglo nombres, para obtener el nÃºmero de veces que se repite nombre.
        contador = nombres.count(nombre)
        if contador > 1 :
            print(f"nombre: {nombre}->{contador}")
```

```python
def cargarDatos():
    with open("test.csv", "r") as data:
        listadoPersonal = []
        for row in data:
            # Asignamos a un arreglo
            campos = row.split(",")
            objeto = dict(ID=campos[0], nombre=campos[1], apellido=campos[2], email=campos[3], valor=campos[4])
            listadoPersonal.append(objeto)
        return listadoPersonal
    
##############################
#1. cargar datos
listadoPersonal = cargarDatos()
busca = input("Ingrese nombre a buscar:")
if(busca != ''):
    # Recorremos la lista
    for persona in listadoPersonal:
        # Comparamos con campo de diccionario
        if persona['nombre']==busca:
            print(f"{persona['apellido']}, {persona['nombre']}")       
        
```
